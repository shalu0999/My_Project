```python

#Import the model
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

import warnings
warnings.filterwarnings("ignore")
import tensorflow as tf
from tensorflow.keras import Sequential 
from tensorflow.keras.layers import Dense 
from sklearn.metrics import classification_report 
```

```python
#Read the dataset
df=pd.read_csv("breast-cancer.csv")
```

```python
#disply top 5 rows of dataset
df.head()
```

```python
#Information about our dataset
df.info()
```

```python
#Check null values 
df.isnull().sum()
```

```python
#Data preprocessing

from sklearn.preprocessing import LabelEncoder 
le = LabelEncoder()

df["diagnosis"]=le.fit_transform(df["diagnosis"])
```

```python
df["diagnosis"].value_counts()
```

```python
df.drop(["id"], axis=1, inplace=True)
```

```python
#Information About Our Dataset
df.info()
```

```python
#Find the relation between the each columns
df.describe()
```

```python
#Store Feature Matrix In X and Target In Vector y
x=df.iloc[:,1:-1]
x
```

```python
y=df["diagnosis"].values
y
```

```python
##spliting the dataset into training set and testings set
from sklearn.model_selection import train_test_split
```

```python
xtrain,xtest,ytrain,ytest=train_test_split(x,y,test_size=0.30,random_state=1)
```

```python
#We use standardscaler
from sklearn.preprocessing import StandardScaler
```

```python
sc=StandardScaler()
```

```python
xtrain=sc.fit_transform(xtrain)
```

```python
xtest=sc.transform(xtest)
```

```python
#import the model
from tensorflow.keras.callbacks import EarlyStopping
early_stop = EarlyStopping(monitor='val_loss', mode='min', verbose=1, patience=10)

#Early stopping:
#1)Early stopping is a method that allows you to specify an arbitrary large no.of training epochs and stop
#training once the model performance stops improving on a hold out validation dataset.
#2)Too many epochs can lead to overfiting of the training dataset, whereas too few may result
#in an underfit model.

#patience:
#number of epochs on which we would like to see no improvement.

#monitor: Quantity to be monitored. by default, it is validation loss

#mode: One of {“auto”, “min”, “max”}. 
#Is it a maximization problem or a minimization problem, we maximize accuracy, minimize loss.

```

```python
##init the model
ann=Sequential()
```

```python
##add a layer in to the model
ann.add(Dense(units=10,activation="relu"))
ann.add(Dense(units=10,activation="relu"))
#Output layer
ann.add(Dense(units=1,activation="sigmoid"))
```

```python
#establish connection between the layers
ann.compile(optimizer="sgd",loss="binary_crossentropy",metrics=["accuracy"])
```

```python
##train model
ann.fit(xtrain,ytrain,epochs=800,validation_data=(xtest,ytest),verbose=1,batch_size=128,callbacks=[early_stop])

#A callback is a powerful tool to customize the behavior of a Keras model during training, 
#evaluation, or inference
```

```python
#Convert the history data in dataframe
loss=pd.DataFrame(ann.history.history)
```

```python
loss
```

```python
##set the threshold ypred=ypred>0.5
ypred=ann.predict(xtest)
ypred=ypred>0.5
#find the accuracy
from sklearn.metrics import classification_report
print(classification_report(ytest,ypred))
```
